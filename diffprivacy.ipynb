```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
df = pd.read_csv("healthdata.csv")
df.head()
```

```python
def laplaceMechanism(x, epsilon):
    x +=  np.random.laplace(0, 1.0/epsilon, 1)[0]
    return x

def gaussianMechanism(x, epsilon):
    x +=  np.random.normal(0, 1.0/epsilon, 1)[0]
    return x
```

```python
df.groupby("problem")['age'].mean().plot(kind='bar')
#df.groupby("insured")['wage'].mean().plot(kind='bar')
```
